```python
from pandas import DataFrame
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

#graph for steady driving behavior with acceleration x
%matplotlib inline
df_normal = pd.read_csv("/Users/dhanushka/Desktop/mydata/fypit/anaconda/data/slow2.csv").iloc[5000:10000]

fig_normal = plt.figure(figsize=(10, 10,))
plt.scatter((df_normal['time']-100).abs(), (df_normal['X']+0.5))
```

```python
#Frequency graph for steady data  acceleration x
pd.cut((df_normal['X']+0.5).abs(), bins=[0,0.5,1,1.5,2,2.5,3,3.5,4,4.5,5,5.5,6]).value_counts().sort_index()
```

```python
# Frequency graph for steady data  acceleration x

plot_data_normal= pd.cut((df_normal['X']+0.5).abs(), bins=[0,0.5,1,1.5,2,2.5,3,3.5,4,4.5,5,5.5,6]).value_counts()
plt.plot([0.25, 0.75, 1.25, 1.75,2.25, 2.75, 3.25, 3.75, 4.25, 4.75, 5.25, 5.75], plot_data_normal)
```

```python
(df_normal['X']+0.5).abs().std()
```

```python
(df_normal['X']+0.5).abs().mean()
```
